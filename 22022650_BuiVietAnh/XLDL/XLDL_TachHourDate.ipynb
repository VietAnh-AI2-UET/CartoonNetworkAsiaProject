```python
import os
import pandas as pd
from datetime import datetime
```

```python
# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c hi·ªán t·∫°i
current_directory = os.getcwd()

# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c cha
parent_directory = os.path.dirname(current_directory)

# T·∫°o ƒë∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß t·ªõi file CSV
csv_path = os.path.join(parent_directory, "CartoonNetworkAsia_81Post_Processing_02_SeperatedReactions.csv")

# ƒê·ªçc file CSV
df = pd.read_csv(csv_path)
```

```python
df
```

```python
# t·∫°o 1 list ch·ª©a c√°c ng√†y th√°ng ƒëƒÉng b√†i
date_list = []

for date in df.time:
    date_list.append(date[:10])
```

```python
# √©p ki·ªÉu thu·ªôc t√≠nh time t·ª´ str th√†nh datetime
i = 0

for time in df.time:
    to_date = datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
    df.iloc[i, 2] = to_date
    i += 1
```

```python
# t·∫°o 1 list ch·ª©a c√°c gi·ªù ƒëƒÉng b√†i
hour_list = []

for time in df.time:
    hour_list.append(time.hour)
```

```python
# t·∫°o dataframe ch·ª©a gi·ªù ƒëƒÉng b√†i
df_hour = pd.DataFrame(hour_list, columns = ["hour"])
```

```python
# t·∫°o dataframe ch·ª©a ng√†y v√† th√°ng ƒëƒÉng b√†i
df_date = pd.DataFrame(date_list, columns = ["date"])
```

```python
# t·∫°o dataframe df_hour_date ch·ª©a gi·ªù, ng√†y th√°ng ƒëƒÉng b√†i
df_hour_date = pd.concat([df_hour, df_date], axis = 1)
```

```python
# b·ªè c·ªôt time trong dataframe c≈©
no_time_df = df.drop(['time'], axis = 1)
```

```python
# n·ªëi dataframe ƒë√£ x√≥a time v·ªõi df_hour_date
new_df = pd.concat([no_time_df, df_hour_date], axis = 1)
```

```python
new_df
```

```python
# t·∫°o list ch·ª©a th·ª© ƒëƒÉng b√†i
day_of_week_list = []

for date in new_df.date:
    specific_date = pd.to_datetime(date)
    day_of_week = specific_date.strftime('%A')
    day_of_week_list.append(day_of_week)
```

```python
# t·∫°o dataframe ch·ª©a th·ª© ƒëƒÉng b√†i
day_df = pd.DataFrame(day_of_week_list, columns = ["day"])
```

```python
day_df
```

```python
# n·ªëi new_df v·ªõi day_df
new_df_2 = pd.concat([new_df, day_df], axis = 1)
```

```python
# l∆∞u file d∆∞·ªõi t√™n CartoonNetworkAsia_81Post_Processing_03_SeperatedHourDate.csv
new_df_2.to_csv("CartoonNetworkAsia_81Post_Processing_03_SeperatedHourDate.csv", index = False)
```
