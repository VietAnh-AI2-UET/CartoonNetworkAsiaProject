```python
import os
import pandas as pd
import ast
```

```python
# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c hi·ªán t·∫°i
current_directory = os.getcwd()

# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c cha
parent_directory = os.path.dirname(current_directory)

# T·∫°o ƒë∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß t·ªõi file CSV
csv_path = os.path.join(parent_directory, "CartoonNetworkAsia_81Post_Processing_01.csv")

# ƒê·ªçc file CSV
df = pd.read_csv(csv_path)
```

```python
df
```

```python
# t·∫°o 1 list ch·ª©a c√°c dictionary c·ªßa c√°c lo·∫°i reaction c·ªßa t·ª´ng post v√† l∆∞u v√†o 1 list 
react_list = []

for react_str in df.reactions:
    react_dict = ast.literal_eval(react_str)
    react_list.append(react_dict)
```

```python
#t·∫°o dataframe t·ª´ react_list
df_reactions = pd.DataFrame(react_list)
```

```python
df_reactions
```

```python
# thay th·∫ø c√°c gi√° tr·ªã NaN b·∫±ng 0
df_reactions = df_reactions.fillna(0)
```

```python
# b·ªè thu·ªôc t√≠nh reactions c≈©
no_react_df = df.drop(['reactions'], axis = 1)
```

```python
# thay thu·ªôc t√≠nh reactions b·∫±ng dataframe df_reactions
new_df = pd.concat([no_react_df, df_reactions], axis = 1)
```

```python
new_df
```

```python
# l∆∞u file d∆∞·ªõi t√™n artoonNetworkAsia_81Post_Processing_02_SeperatedReactions.csv
new_df.to_csv("CartoonNetworkAsia_81Post_Processing_02_SeperatedReactions.csv", index = False)
```
