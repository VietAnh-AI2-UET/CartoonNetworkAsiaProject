```python
# c√°c ch∆∞∆°ng tr√¨nh ƒë∆∞·ª£c ƒëƒÉng b√†i
show_dict = {
    'tom and jerry': len(taj),
    'Regular Show': len(rgs),
    'Teen Titans Go': len(ttg),
    'We Bare Bears': len(wbb),
    'Ben 10': len(b10),
    'Adventure Time': len(adt),
    'Looney Tunes Cartoons': len(ltc),
    'The Heroic Quest Of The Valiant Prince Ivandoe': len(thqotvpi),
    'Craig of the Creek': len(cotc),
    'Lamput': len(lp),
    'Stream your Cartoon Network favorites on HBO Asia': len(sycnfoha),
    'Batman Day on Cartoon Network': len(bmdocn),
    'Catch your favorite Cartoon Network shows': len(cyfcns),
    'Cat-stravaganza': len(csvz)
}
```

```python
# t·ªâ tr·ªçng s·ªë b√†i ƒëƒÉng gi·ªØa c√°c ch∆∞∆°ng tr√¨nh
show = []
post = []

explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.2, 0.4, 0.3, 0.2, 0.5, 0.2, 0.0, 0.1, 0.0)

for x, y in show_dict.items():
    show.append(x)
    post.append(y)

plt.pie(post, labels=['']*len(show), autopct='%.2f', explode = explode, startangle = 90)

plt.axis('equal')
plt.title("T·ªâ tr·ªçng s·ªë l∆∞·ª£ng b√†i ƒëƒÉng v·ªÅ c√°c ch∆∞∆°ng tr√¨nh")

plt.legend(show, title="Shows", loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
show_dict
```

```python
# s·ªë b√†i ƒëƒÉng c·ªßa t·ª´ng ch∆∞∆°ng tr√¨nh
plt.barh(show, post)

for i, v in enumerate(post):
    plt.text(v + 0.5, i, str(v), color='black', va='center')

plt.xlim(0, max(post) + 2)
plt.show()
```

```python
# reaction c√°c b√†i ƒëƒÉng c·ªßa t·ª´ng ch∆∞∆°ng tr√¨nh
reaction_dict = {
    'tom and jerry': sum(taj.reaction_count),
    'Regular Show': sum(rgs.reaction_count),
    'Teen Titans Go': sum(ttg.reaction_count),
    'We Bare Bears': sum(wbb.reaction_count),
    'Ben 10': sum(b10.reaction_count),
    'Adventure Time': sum(adt.reaction_count),
    'Looney Tunes Cartoons': sum(ltc.reaction_count),
    'The Heroic Quest Of The Valiant Prince Ivandoe': sum(thqotvpi.reaction_count),
    'Craig of the Creek': sum(cotc.reaction_count),
    'Lamput': sum(lp.reaction_count),
    'Stream your Cartoon Network favorites on HBO Asia': sum(sycnfoha.reaction_count),
    'Batman Day on Cartoon Network': sum(bmdocn.reaction_count),
    'Catch your favorite Cartoon Network shows': sum(cyfcns.reaction_count),
    'Cat-stravaganza': sum(csvz.reaction_count)
}
```

```python
# reaction c√°c b√†i ƒëƒÉng c·ªßa t·ª´ng ch∆∞∆°ng tr√¨nh
reaction_dict_2 = {
    'tom and jerry': sum(taj.reaction_count),
    'Regular Show': sum(rgs.reaction_count),
    'We Bare Bears': sum(wbb.reaction_count),
    'Adventure Time': sum(adt.reaction_count),
    'Others': sum(ttg.reaction_count) + sum(b10.reaction_count) + sum(ltc.reaction_count)
                + sum(thqotvpi.reaction_count) + sum(cotc.reaction_count) + sum(lp.reaction_count) + sum(sycnfoha.reaction_count)
                + sum(cyfcns.reaction_count) + sum(csvz.reaction_count) + sum(bmdocn.reaction_count)
}
```

```python
# t·ªâ tr·ªçng s·ªë l∆∞·ª£ng reaction gi·ªØa c√°c ch∆∞∆°ng tr√¨nh
show = []
reaction = []

explode = (0.1, 0.0, 0.2, 0.2, 0.0)

for x, y in reaction_dict_2.items():
    show.append(x)
    reaction.append(y)

plt.pie(reaction, labels=['']*len(show), autopct='%.2f', explode = explode, startangle = 90)

plt.axis('equal')
plt.title("T·ªâ tr·ªçng s·ªë l∆∞·ª£ng reaction gi·ªØa c√°c ch∆∞∆°ng tr√¨nh")

plt.legend(show, title="Shows", loc="center left", bbox_to_anchor=(1, 0.5))
```

```python
# s·ªë l∆∞·ª£ng reaction c·ªßa t·ª´ng ch∆∞∆°ng tr√¨nh
show = []
reaction = []

for x, y in reaction_dict.items():
    show.append(x)
    reaction.append(y)

plt.barh(show, reaction)

for i, v in enumerate(reaction):
    plt.text(v + 0.5, i, str(v), color='black', va='center')

plt.xlim(0, 1.125 * max(reaction))
plt.show()
```
