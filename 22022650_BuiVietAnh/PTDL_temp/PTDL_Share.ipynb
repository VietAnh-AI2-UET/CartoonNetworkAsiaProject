```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c hi·ªán t·∫°i
current_directory = os.getcwd()

# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c cha
parent_directory = os.path.dirname(current_directory)

# T·∫°o ƒë∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß t·ªõi file CSV
csv_path = os.path.join(parent_directory, "Processed\\CartoonNetworkAsia_81Post_Processed.csv")

# ƒê·ªçc file CSV
df = pd.read_csv(csv_path)
```

```python
# th√™m index cho dataframe
df = df.reset_index(inplace = False)
df["index"] = df["index"] + 1
```

```python
df
```

```python
# top 10 b√†i ƒëƒÉng c√≥ l∆∞·ª£t share nhi·ªÅu nh·∫•t
df.sort_values(by = ['shares'], ascending = False).head(10)
```

```python
# bi·ªÉu di·ªÖn b·∫±ng bi·ªÉu ƒë·ªì
top_share = df.sort_values(by = ['shares'], ascending = False).head(10)
top_share_plt = top_share.plot(kind = "bar", x = "index", y = "shares")

for b in top_share_plt.patches:
    top_share_plt.annotate(str(b.get_height()), (b.get_x() + b.get_width() / 2., b.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# gi·ªù ƒëƒÉng b√†i c·ªßa 10 b√†i c√≥ nhi·ªÅu share nh·∫•t
top_share = df.sort_values(by = ['shares'], ascending = False).head(10)
top_share_plt = top_share.plot(kind = "bar", x = "index", y = "hour")

for b in top_share_plt.patches:
    top_share_plt.annotate(str(b.get_height()), (b.get_x() + b.get_width() / 2., b.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# Quan h·ªá gi·ªØa gi·ªù ƒëƒÉng b√†i v√† share
share_hour = df[["shares", "hour"]].groupby(["hour"]).sum("shares")
share_hour_plt = share_hour.plot(kind = "bar")
for b in share_hour_plt.patches:
    share_hour_plt.annotate(str(b.get_height()), (b.get_x() + b.get_width() / 2., b.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# N·ªôi dung c·ªßa top c√°c b√†i ƒëƒÉng
df.sort_values(by = ['shares'], ascending = False).head(10)[["index", "post_text"]]
```

```python
# ng√†y ƒëƒÉng b√†i
df.sort_values(by = ['shares'], ascending = False).head(10)[['index', 'date']]
```

```python
# quan h·ªá c·ªßa s·ªë l∆∞·ª£ng share v√† ng√†y ƒëƒÉng b√†i
top_share = df.sort_values(by = ['shares'], ascending = False).head(10)
top_share_date = top_share.sort_values(by = ['date'], ascending = True)
top_share_date_plt = top_share_date.plot(kind = 'bar', x = 'date', y = 'shares')
i = 0

for b in top_share_date_plt.patches:
    top_share_date_plt.annotate("(" + str(top_share_date.iloc[i]["index"]) + ")\n" + str(b.get_height()), (b.get_x() + b.get_width() / 2., b.get_height()), ha='center', va='center', xytext=(0, 12), textcoords='offset points')
    i += 1

extra_space = 0.15
plt.ylim(top=top_share_date["shares"].max() * (1 + extra_space))

plt.show()
```
