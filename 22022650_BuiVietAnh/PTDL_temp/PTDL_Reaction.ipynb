```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c hi·ªán t·∫°i
current_directory = os.getcwd()

# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c cha
parent_directory = os.path.dirname(current_directory)

# T·∫°o ƒë∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß t·ªõi file CSV
csv_path = os.path.join(parent_directory, "Processed\\CartoonNetworkAsia_81Post_Processed.csv")

# ƒê·ªçc file CSV
df = pd.read_csv(csv_path)
```

```python
# th√™m index cho dataframe
df = df.reset_index(inplace = False)
df["index"] = df["index"] + 1
```

```python
df
```

```python
# 10 b√†i c√≥ l∆∞·ª£ng reaction cao nh·∫•t
df.sort_values(by = ['reaction_count'], ascending = False).head(10)
```

```python
# bi·ªÉu di·ªÖn b·∫±ng bi·ªÉu ƒë·ªì
top_react = df.sort_values(by = ['reaction_count'], ascending = False).head(10)
top_react_plt = top_react.plot(kind = "bar", x = 'index', y = 'reaction_count')

for bar in top_react_plt.patches:
    top_react_plt.annotate(str(bar.get_height()), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

```python
# gi·ªù ƒëƒÉng b√†i
top_hour = df.sort_values(by = ['reaction_count'], ascending = False).head(10).plot(kind = 'bar', x = 'index', y = 'hour')

for bar in top_hour.patches:
    top_hour.annotate(str(bar.get_height()), (bar.get_x() + bar.get_width() / 2., bar.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# Quan h·ªá gi·ªØa gi·ªù ƒëƒÉng b√†i v√† reaction
react_hour = df[["reaction_count", "hour"]].groupby(["hour"]).sum("reaction_count")
react_hour_plt = react_hour.plot(kind = "bar")
for b in react_hour_plt.patches:
    react_hour_plt.annotate(str(b.get_height()), (b.get_x() + b.get_width() / 2., b.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# N·ªôi dung c·ªßa top c√°c b√†i ƒëƒÉng
df.sort_values(by = ['reaction_count'], ascending = False).head(10)[["index", "post_text"]]
```

```python
# ng√†y ƒëƒÉng b√†i
df.sort_values(by = ['reaction_count'], ascending = False).head(10)[['index', 'date']]
```

```python
# quan h·ªá c·ªßa s·ªë l∆∞·ª£ng reaction v√† ng√†y ƒëƒÉng b√†i
top_react = df.sort_values(by = ['reaction_count'], ascending = False).head(10)
top_react_date = top_react.sort_values(by = ['date'], ascending = True)
top_react_date_plt = top_react_date.plot(kind = 'bar', x = 'date', y = 'reaction_count')
i = 0

for bar in top_react_date_plt.patches:
    top_react_date_plt.annotate("(" + str(top_react_date.iloc[i]["index"]) + ")\n" + str(bar.get_height()), (bar.get_x() + bar.get_width() / 2., bar.get_height()), ha='center', va='center', xytext=(0, 12), textcoords='offset points')
    i += 1

extra_space = 0.15 
plt.ylim(top=top_react_date["reaction_count"].max() * (1 + extra_space))

plt.show()
```
