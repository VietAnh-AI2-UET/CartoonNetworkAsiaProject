```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c hi·ªán t·∫°i
current_directory = os.getcwd()

# L·∫•y ƒë∆∞·ªùng d·∫´n c·ªßa th∆∞ m·ª•c cha
parent_directory = os.path.dirname(current_directory)

# T·∫°o ƒë∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß t·ªõi file CSV
csv_path = os.path.join(parent_directory, "Processed\\CartoonNetworkAsia_81Post_Processed.csv")

# ƒê·ªçc file CSV
df = pd.read_csv(csv_path)
```

```python
# th√™m index cho dataframe
df = df.reset_index(inplace = False)
df["index"] = df["index"] + 1
```

```python
df
```

```python
# th·ªùi gian hay ƒëƒÉng b√†i
hour_plt = df.hour.value_counts().plot(kind = "bar", ylabel = 'frequency')

for b in hour_plt.patches:
    hour_plt.annotate(b.get_height(), (b.get_x() + b.get_width() / 2, b.get_height()), ha = 'center', va = 'center', xytext = (0,5), textcoords = 'offset points')

plt.show()
```

```python
# bi·∫øn ƒë·ªông th·ªùi gian ƒëƒÉng b√†i
df.plot(kind = "line", x = "index", y = "hour", drawstyle='steps')
```

```python
# chuy·ªÉn ƒë·ªãnh d·∫°ng thu·ªôc t√≠nh date th√†nh datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# s·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo t·ª´ng ng√†y
df['date'].value_counts().sort_index().plot(kind='bar', figsize=(50, 20))
```

```python
# S·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo t·ª´ng ng√†y
df.date.value_counts().plot(kind = "line", drawstyle = "steps", grid = True, figsize = (15, 7))
```
